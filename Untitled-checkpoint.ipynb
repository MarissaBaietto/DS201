```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('real_estate_data_DS201.csv')
```

How are we going to "clean" this dataset?

```python
df.head()
```

Create a new column that shows how old the house is (assuming this data is from 2017)

```python
df['house_age'] = 2017 - df['year_built']
```

```python
df['house_age']
```

Create a scatterplot comparing the house age and price.

```python
sns.scatterplot(data=df, x=df['house_age'], y=df['price_closed'], hue=None, size=None, style=None, palette=None)
```

Create a histogram of prices to see distribution.

```python
plt.hist(df['price_closed'], bins=20, log=False)
plt.ticklabel_format(style='plain')
plt.xticks(rotation=45)
plt.show()
```

Create a pairplot showing bathrooms, bedrooms, and price.

```python
sns.pairplot(
    df,
    diag_kind="hist",
    x_vars=["bedrooms_total", "baths_total", "price_closed"],
    y_vars=["bedrooms_total", "baths_total", "price_closed"],
)
sns.set_theme(style='white')
```
